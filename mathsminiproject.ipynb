## importing the necessary packages

```python
import networkx as nx
import matplotlib.pyplot as plt
G = nx.Graph()
```

### adding nodes to the network

```python
G.add_nodes_from(["john","steve","alex","jack","rose","luke","chris","karl","jacob","dominic","eric","tony","mark"])
```

### adding edges between nodes to signify their relation

```python
G.add_edges_from([("john","steve"),("steve","alex"),("jack","rose"),("chris","karl"),("jacob","dominic"),("steve","chris")])
G.add_edges_from([("alex","dominic"),("steve","karl"),("jack","luke"),("chris","jacob"),("jacob","rose"),("steve","jacob")])
G.add_edges_from([("john","luke"),("eric","alex"),("alex","rose"),("eric","karl"),("john","dominic"),("jack","karl")])
G.add_edges_from([("tony","mark"),("tony","alex"),("mark","rose"),("chris","luke"),("eric","dominic"),("rose","chris")])
```

## visualising the relationship between the nodes

```python
nx.draw(G,with_labels=True)
```
